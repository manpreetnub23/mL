```python
choice = int(input(prompt = '1 for 1d array. /n 2 for 2d array.'))
n = int(input(prompt = 'enter number of rows : '))
if choice == 2:
    m = int(input(prompt = 'enter number of cols : '))
    batao1 = int(input('1 for array with zeros. /n 2 for array with ones. /n 3 for random array.'))
    if batao1 == 1:
        array1 = np.zeros((n,m), dtype = 'int64')
        print(array1)
    elif batao1 == 2:
        array2 = np.ones((n,m), dtype = 'int64')
        print(array2)
    else:
        array3 = np.random.randint(1,10,(n,m))
        print(array3)
else:
    batao2 = int(input('1 for array with zeros. /n 2 for array with ones. /n 3 for random array.'))
    if batao2 == 1:
        onedarray1 = np.zeros((n), dtype = 'int64')
        print(onedarray1)
    elif batao2 == 2:
        onedarray2 = np.ones((n), dtype = 'int64')
        print(onedarray2)
    else:
        onedarray3 = np.random.randint(1,10,(n))
        print(onedarray3)
```
