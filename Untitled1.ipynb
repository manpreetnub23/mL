```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('fuel.csv')
```

```python
for i in df.columns:
     print("{0:.^7}".format(i[0:5]),end='|')
print()
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        t = type(df.iloc[i,j])
        if t == str:
            print(f"{str(df.iloc[i][j])[0:7]:.<7}|",end= '')
        elif t == np.int64:
            print(f"{str(df.iloc[i][j])[0:7]:.^7}|",end= '')
        else:
            print(f"{str(df.iloc[i][j])[0:7]:.>7}|",end= '')
    print()
```
